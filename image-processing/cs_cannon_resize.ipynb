```r
image_resize_smw <- function(i, clobber = FALSE, resize = 52) {
    i.newfile <- paste0(str_replace(i, "/HW/", paste0("/", folder.name, "/")), ".jpg")
    if (file.exists(i.newfile) && !clobber) {
        message("File already exists:", i.newfile)
    } else {
        i.read <- image_read(i)
        i.new <- image_resize(i.read, geometry_size_percent(resize)) # TODO: 30%
        # i.new <- image_resize(i.read, geometry_size_pixels(2000, 1333))
        image_write(i.new, path = i.newfile)
        message("Wrote new file", i.newfile)
        
        # Cleanup, to avoid memory errors
        image_destroy(i.read)
        image_destroy(i.new)
        rm(i.read, i.new)
    }
    i.newfile
}
```

```r
mclapply(files.cannon, image_resize_smw, mc.cores = parallel::detectCores())
```
